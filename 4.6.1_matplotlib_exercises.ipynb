```python
#  4.6 exercises in matlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as math
x = list(range(-5,7))
y = [ (n**2 - n + 2) for n in x]
plt.plot(x,y)
# plt.annotate('(0,0)', xy=(0, 0), xytext=(0, 150), arrowprops={'facecolor': 'red'})
```

Create and label 4 separate plots for the following equations (choose a range for x that makes sense):

y = âˆšx
y = x^3
y = tan(x)
y = 2^x

You can use plt.figure to set the size of the plots, and hspace gives some margin between the upper and lower plots

```python
# 4 subplots
n_rows = 2
n_cols = 2
x = list(range(0,10))
plt.figure(figsize=(7, 6)).subplots_adjust(hspace=0.4)
# sub plot 1
plt.subplot(n_rows, n_cols, 1)
y = [math.sqrt(n) for n in x]
plt.title('y = sqrt of x')
plt.plot(x, y, c='orange')
# plot 2
x = list(range(-10,10))
plt.subplot(n_rows, n_cols, 2)
y = [(n**3) for n in x]
plt.title('y = x cubed')
plt.plot(x, y, c='green')
# plot 3
x = list(range(0,10))
plt.subplot(n_rows, n_cols, 3)
y = [math.tan(n) for n in x]
plt.title('y = tangent of x')
plt.plot(x, y, c = 'red')
# plot 4
x = list(range(-10,10))
plt.subplot(n_rows, n_cols, 4)
y = [(2**n) for n in x]
plt.title('y = 2 to x power')
plt.plot(x,y, c = 'purple')
plt.suptitle('4 different functions of y(x)')
plt.show()
```

Combine 4 separate functions to single chart (choose a range for x that makes sense):

y = âˆšx
y = x^3
y = tan(x)
y = 2^x

```python
plt.figure(figsize=(7, 7))
x = list(range(1,10))
z1 = [math.sqrt(n) for n in x]
plt.plot(x, z1, c = 'orange', label='y = sqrt of x')
z2 = [(n**3) for n in x]
plt.plot(x, z2, c = 'green', label='y = x cubed')
z3 = [math.tan(n) for n in x]
plt.plot(x, z3, c = 'red', label='y = tangent of x')
z4 = [(2**n) for n in x]
plt.plot(x,z4, c = 'purple', label='y = 2 to x power')
plt.legend(loc='upper left')
plt.suptitle('Four combined functions')
plt.show()
```
